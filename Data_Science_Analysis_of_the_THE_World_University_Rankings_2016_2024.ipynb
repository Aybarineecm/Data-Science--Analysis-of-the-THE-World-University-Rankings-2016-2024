```python
import pandas as pd
```

```python
df=pd.read_csv('THE World University Rankings 2016-2024.csv', encoding= "latin1")
```

Displaying the first five rows of the dataset.

```python
df.head(5)
```

Displaying the last five rows of the dataset.

```python
df.tail(5)
```

Showing how many rows have missing values.

```python
df.isnull()
```

 Removing any row that has missing values.

```python
df.dropna()
```

Renaming the column Student Population to Population

```python
df= df.rename(columns={'Student Population': 'Population'})
```

Display the country and name of all the universities that have more than 20,000 students

```python
df.head()
```

Display the country and name of all the universities that have more than 20,000 students

```python
df[

    df['Population']>20000][['Country', 'Name', 'Population']

    ]
```

Filter all the rows that have Student to Staff ratio higher than 10 and an overall score greater than 60.

```python
df[
     (df['Students to Staff Ratio']>10) & (df['Overall Score']>60)

     ]
```

Show the average student population grouped by country

```python
df.groupby(['Country'])['Population'].mean()
```

Show the total number of students grouped by country

```python
df.groupby(['Country'])['Population'].sum()
```

Find the university with the max overall score by country

```python
df.groupby(['Country'])['Overall Score'].max()
```

Show the average student to staff ratio grouped by country

```python
df.groupby(['Country'])['Students to Staff Ratio'].mean()
```
